```python
from sklearn.utils import resample
majority_class = df[df['FraudFound'] == 0]
minority_class = df[df['FraudFound'] == 1]
undersampled_size = minority_class.shape[0]
undersampled_majority = resample(majority_class,replace=False,n_samples=undersampled_size,random_state=42)
undersampled_df = pd.concat([undersampled_majority, minority_class])
undersampled_df = undersampled_df.sample(frac=1, random_state=42).reset_index(drop=True)
undersampled_class_counts = undersampled_df['FraudFound'].value_counts()
print(undersampled_class_counts)
```

```python
undersampled_df
```

```python
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,f1_score
```

```python
X = undersampled_df.drop('FraudFound',axis=1)
y = undersampled_df['FraudFound']
```

```python
model = svm.SVC(kernel="linear")
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print("Accuracy is : ", accuracy_score(y_pred, y_test))
print("Precision is : ", precision_score(y_pred, y_test, average="weighted"))
print("F1 score is : ", f1_score(y_pred, y_test, average="weighted"))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr = LinearRegression()
lr.fit(X_train, y_train)
y_predlr = lr.predict(X_test)
mse = mean_squared_error(y_test, y_predlr)
r2 = r2_score(y_test, y_predlr)
print(f"Mean Squared Error: {mse}")
print(f"R-squared Score: {r2}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_predknn = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_predknn)
print("Accuracy is : ", accuracy_score(y_predknn, y_test))
print("Precision is : ", precision_score(y_predknn, y_test))
print("F1 score is : ", f1_score(y_predknn, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
dt=DecisionTreeClassifier(criterion='gini',max_leaf_nodes=10).fit(X_train, y_train)
plt.figure(figsize=(10, 10))
plot_tree(dt, feature_names=X_train.columns, class_names=['0', '1'], filled=True)
plt.show()
y_preddt = dt.predict(X_test)
accuracy = accuracy_score(y_test, y_preddt)
print(f"Accuracy: {accuracy}")
```
